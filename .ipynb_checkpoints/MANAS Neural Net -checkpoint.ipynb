```python
def get_image_data(path,classes):
    X=[]
    Y=[]
    for cls in classes:
        cl_path = path + "/" +cls
        file_list = os.listdir(cl_path)
        for image in file_list:
            im = np.array(Image.open(cl_path + "/" + image))
            X.append(im)
            Y.append(cls)
    return np.array(X),np.array(Y)
```

```python
#Let's get the easiest thing done first
def sigmoid(z):
    return 1/(1+np.exp(-z))

def BinaryCrossEntropy(a):
    cost = -np.mean(a*np.log(a)+(1-a)*np.log(1-a))
    return cost
```

## Learning Algorithm: Google tells me that Neural Networks also use variations of the same techniques that we used to train our Logisitic Regression Model , Gradient Descent. Now there are three types of Gradient Descent:
### 1. Batch Gradient Descent : The derivatives are calculated for each set of weights in the network , and the weights are changed only once all the samples have been trained on. Although it produces a relatively stable convergence, it is not often the best and it also requires the entire dataset to be available to the network in memory at all times.
### 2. Stochastic Gradient Descent :  Here, instead of waiting for the world to end, the weights are updated based on every loss and derivative calculation of every sample. Since it doesn't require the whole data to be used for updation at once, it is faster but computationaly very expensive given the frequency of updates
### 3. Mini-Batch Gradient Descent : Taking the best of both worlds, Mini-Batch gradient descent, updates all the weights every time after training the model on a fixed number(mini-batch) of training samples and then trains the model on those updated set of weights. 

#### So for my model, I am planning to use min-batch-Stochastic Gradient Descent

## Alright, let's put them together  in a class with several methods

```python
import random
import numpy as np

class Network(object):

    def __init__(self, sizes):
        self.num_layers = len(sizes)
        self.sizes = sizes
        self.biases = [np.random.randn(y, 1) for y in sizes[1:]]
        self.weights = [np.random.randn(y, x)
                        for x, y in zip(sizes[:-1], sizes[1:])]

    def feedforward(self, a):
        for b, w in zip(self.biases, self.weights):
            a = sigmoid(np.dot(w, a)+b)
        return a

    def SGD(self, training_data, epochs, mini_batch_size, lr,
            test_data=None):
        if test_data: n_test = len(test_data)
        n = len(training_data)
        for j in range(epochs):
            random.shuffle(training_data)
            mini_batches = [
                training_data[k:k+mini_batch_size]
                for k in range(0, n, mini_batch_size)]
            for mini_batch in mini_batches:
                self.update_mini_batch(mini_batch, lr)
            if test_data:
                tr,te = self.evaluate(test_data)
                print("Epoch",j,"Complete")
                print("On Training Set:",tr,"/",n)
                print("On Test Set:",te,"/",n_test)
            else:
                print("Epoch {0} complete".format(j))
            print("Weight",self.weights[-1][0][:10])

    def update_mini_batch(self, mini_batch, lr):
        nabla_b = [np.zeros(b.shape) for b in self.biases]
        nabla_w = [np.zeros(w.shape) for w in self.weights]
        for x, y in mini_batch:
            x=x.reshape(784,1)
            delta_nabla_b, delta_nabla_w = self.backprop(x, y)
            nabla_b = [nb+dnb for nb, dnb in zip(nabla_b, delta_nabla_b)]
            nabla_w = [nw+dnw for nw, dnw in zip(nabla_w, delta_nabla_w)]
        self.weights = [w-(lr/len(mini_batch))*nw
                        for w, nw in zip(self.weights, nabla_w)]
        self.biases = [b-(lr/len(mini_batch))*nb
                       for b, nb in zip(self.biases, nabla_b)]

    def backprop(self, x, y):
        nabla_b = [np.zeros(b.shape) for b in self.biases]
        nabla_w = [np.zeros(w.shape) for w in self.weights]
        activation = x
        activations = [x] # list to store all the activations, layer by layer
        zs = [] # list to store all the z vectors, layer by layer
        for b, w in zip(self.biases, self.weights):
            z = np.dot(w, activation)+b
            zs.append(z)
            activation = sigmoid(z)
            activations.append(activation)
        # backward pass
        delta = self.cost_derivative(activations[-1], y.reshape(10,1)) * \
            sigmoid_prime(zs[-1])
        nabla_b[-1] = delta
        nabla_w[-1] = np.dot(delta, activations[-2].transpose())
 
        for l in range(2, self.num_layers):
            z = zs[-l]
            sp = sigmoid_prime(z)
            delta = np.dot(self.weights[-l+1].transpose(), delta) * sp
            nabla_b[-l] = delta
            nabla_w[-l] = np.dot(delta, activations[-l-1].transpose())
        return (nabla_b, nabla_w)

    def evaluate(self, test_data):
        test_results = [(np.argmax(self.feedforward(x.reshape(784,1))), y)
                        for (x, y) in test_data]
        train_results = [(np.argmax(self.feedforward(x.reshape(784,1))),y)
                         for (x,y) in zip(X_train_std,Y_train_numeric)]
     
        train_sum = sum(int(x==y) for (x,y) in train_results)
        test_sum = sum(int(x==y) for (x,y) in test_results)
        return train_sum,test_sum

    def cost_derivative(self, output_activations, y):

        return (output_activations-y)

def sigmoid(z):
    """The sigmoid function."""
    return 1.0/(1.0+np.exp(-z))

def sigmoid_prime(z):
    """Derivative of the sigmoid function."""
    return sigmoid(z)*(1-sigmoid(z))
```

## Let us work around that, one way google suggests me to handle it is to use something called regularization, that keeps the weights in check and prevents them from learning the data just too well.

```python
import random

# Third-party libraries
import numpy as np

class Network4(object):

    def __init__(self, sizes):
        self.num_layers = len(sizes)
        self.sizes = sizes
        self.biases = [np.random.randn(y, 1) for y in sizes[1:]]
        self.weights = [np.random.randn(y, x)
                        for x, y in zip(sizes[:-1], sizes[1:])]

    def feedforward(self, a):
        for b, w in zip(self.biases, self.weights):
            a = sigmoid(np.dot(w, a)+b)
        return a

    def SGD(self, training_data, epochs, mini_batch_size, lr,
            lmbda = 0.005 ,reg=True, test_data=None):
        if test_data: n_test = len(test_data)
        n = len(training_data)
        for j in range(epochs):
            random.shuffle(training_data)
            mini_batches = [
                training_data[k:k+mini_batch_size]
                for k in range(0, n, mini_batch_size)]
            for mini_batch in mini_batches:
                self.update_mini_batch(mini_batch, lr , lmbda , reg)
            if test_data:
                tr,te = self.evaluate(test_data)
                print("Epoch",j,"Complete")
                print("On Training Set:",tr,"/",n)
                print("On Test Set:",te,"/",n_test)
            else:
                print("Epoch {0} complete".format(j))
            print("Weight",self.weights[-1][0][:10])

    def update_mini_batch(self, mini_batch, lr , lmbda = 0.005,reg =True):
        if(not reg):
            lmbda = 0
        #Here the lmbda term that I have added is called the regularization parameter and at present I am using L2 Regularization
        #which means adding the sum of squares of coefficients(here weights) to the Loss Function 
        
        nabla_b = [np.zeros(b.shape) for b in self.biases]
        nabla_w = [np.zeros(w.shape) for w in self.weights]
        for x, y in mini_batch:
            x=x.reshape(784,1)
            delta_nabla_b, delta_nabla_w = self.backprop(x, y)
            nabla_b = [nb+dnb for nb, dnb in zip(nabla_b, delta_nabla_b)]
            nabla_w = [nw+dnw for nw, dnw in zip(nabla_w, delta_nabla_w)]
        self.weights = [w-(lr/len(mini_batch))*nw-2*lmbda*w 
                        for w, nw in zip(self.weights, nabla_w)]
        self.biases = [b-(lr/len(mini_batch))*nb
                       for b, nb in zip(self.biases, nabla_b)]

    def backprop(self, x, y):
        nabla_b = [np.zeros(b.shape) for b in self.biases]
        nabla_w = [np.zeros(w.shape) for w in self.weights]
        activation = x
        activations = [x] # list to store all the activations, layer by layer
        zs = [] # list to store all the z vectors, layer by layer
        for b, w in zip(self.biases, self.weights):
            z = np.dot(w, activation)+b
            zs.append(z)
            activation = sigmoid(z)
            activations.append(activation)
        # backward pass
        delta = self.cost_derivative(activations[-1], y.reshape(10,1)) * \
            sigmoid_prime(zs[-1])
        nabla_b[-1] = delta
        nabla_w[-1] = np.dot(delta, activations[-2].transpose())
 
        for l in range(2, self.num_layers):
            z = zs[-l]
            sp = sigmoid_prime(z)
            delta = np.dot(self.weights[-l+1].transpose(), delta) * sp
            nabla_b[-l] = delta
            nabla_w[-l] = np.dot(delta, activations[-l-1].transpose())
        return (nabla_b, nabla_w)

    def evaluate(self, test_data):
        test_results = [(np.argmax(self.feedforward(x.reshape(784,1))), y)
                        for (x, y) in test_data]
        train_results = [(np.argmax(self.feedforward(x.reshape(784,1))),y)
                         for (x,y) in zip(X_train_std,Y_train_numeric)]
     
        train_sum = sum(int(x==y) for (x,y) in train_results)
        test_sum = sum(int(x==y) for (x,y) in test_results)
        return train_sum,test_sum

    def cost_derivative(self, output_activations, y):

        return (output_activations-y)


def sigmoid(z):
    """The sigmoid function."""
    return 1.0/(1.0+np.exp(-z))

def sigmoid_prime(z):
    """Derivative of the sigmoid function."""
    return sigmoid(z)*(1-sigmoid(z))
```

```python
import random

# Third-party libraries
import numpy as np

class Network(object):

    def __init__(self, sizes):
        self.num_layers = len(sizes)
        self.sizes = sizes
        self.biases = [np.random.randn(y, 1) for y in sizes[1:]]
        self.weights = [np.random.randn(y, x)
                        for x, y in zip(sizes[:-1], sizes[1:])]
        self.X = None

    def feedforward(self, a):
        for b, w in zip(self.biases, self.weights):
            a = sigmoid(np.dot(w, a)+b)
        return a

    def SGD(self, training_data, epochs, mini_batch_size, lr,
            lmbda = 0.005 ,reg=True, test_data=None):
        if test_data: 
            n_test = len(test_data)
        n = len(training_data)
        self.X = training_data
        for j in range(epochs):
            random.shuffle(training_data)
            mini_batches = [
                training_data[k:k+mini_batch_size]
                for k in range(0, n, mini_batch_size)]
            for mini_batch in mini_batches:
                self.update_mini_batch(mini_batch, lr , lmbda , reg)
            if test_data:
                tr,te = self.evaluate(test_data)
                print("On Training Set:",tr,"/",n)
                print("On Test Set:",te,"/",n_test)
            else:
                print("Epoch {0} complete".format(j))
            print("Weight",self.weights[-1][0][:10])

    def update_mini_batch(self, mini_batch, lr , lmbda = 0.005,reg =True):
        if(not reg):
            lmbda = 0
        #Here the lmbda term that I have added is called the regularization parameter and at present I am using L2 Regularization
        #which means adding the sum of squares of coefficients(here weights) to the Loss Function 
        
        nabla_b = [np.zeros(b.shape) for b in self.biases]
        nabla_w = [np.zeros(w.shape) for w in self.weights]
        for x, y in mini_batch:
            x=x.reshape(784,1)
            delta_nabla_b, delta_nabla_w = self.backprop(x, y)
            nabla_b = [nb+dnb for nb, dnb in zip(nabla_b, delta_nabla_b)]
            nabla_w = [nw+dnw for nw, dnw in zip(nabla_w, delta_nabla_w)]
        self.weights = [w-(lr/len(mini_batch))*nw-2*lmbda*w 
                        for w, nw in zip(self.weights, nabla_w)]
        self.biases = [b-(lr/len(mini_batch))*nb
                       for b, nb in zip(self.biases, nabla_b)]

    def backprop(self, x, y):
        nabla_b = [np.zeros(b.shape) for b in self.biases]
        nabla_w = [np.zeros(w.shape) for w in self.weights]
        activation = x
        activations = [x] # list to store all the activations, layer by layer
        zs = [] # list to store all the z vectors, layer by layer
        for b, w in zip(self.biases, self.weights):
            z = np.dot(w, activation)+b
            zs.append(z)
            activation = sigmoid(z)
            activations.append(activation)
        # backward pass
        delta = self.cost_derivative(activations[-1], y.reshape(10,1)) * \
            sigmoid_prime(zs[-1])
        nabla_b[-1] = delta
        nabla_w[-1] = np.dot(delta, activations[-2].transpose())
 
        for l in range(2, self.num_layers):
            z = zs[-l]
            sp = sigmoid_prime(z)
            delta = np.dot(self.weights[-l+1].transpose(), delta) * sp
            nabla_b[-l] = delta
            nabla_w[-l] = np.dot(delta, activations[-l-1].transpose())
        return (nabla_b, nabla_w)

    def evaluate(self, test_data):
        test_results = [(np.argmax(self.feedforward(x.reshape(784,1))), y)
                        for (x, y) in test_data]
        train_results = [(np.argmax(self.feedforward(x.reshape(784,1))),np.argmax(y))
                         for (x,y) in self.X]
     
        train_sum = sum(int(x==y) for (x,y) in train_results)
        test_sum = sum(int(x==y) for (x,y) in test_results)
        return train_sum,test_sum

    def cost_derivative(self, output_activations, y):

        return (output_activations-y)


def sigmoid(z):
    """The sigmoid function."""
    return 1.0/(1.0+np.exp(-z))

def sigmoid_prime(z):
    """Derivative of the sigmoid function."""
    return sigmoid(z)*(1-sigmoid(z))
```
