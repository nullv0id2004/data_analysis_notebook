```python
def find_outliers_IQR(d):
    q1=d.quantile(0.25)

    q3=d.quantile(0.75)

    IQR=q3-q1

    outliers = d[((d<(q1-1.5*IQR)) | (d>(q3+1.5*IQR)))]

    return outliers

for column in temp[['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']]:
    outliers = find_outliers_IQR(temp[column])
    temp1 = temp.replace(outliers,temp[column].median())
temp1
```
