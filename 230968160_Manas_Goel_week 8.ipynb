## 1. b. Select only the transactions that have occurred from 01/04/ 2011 and 09/12/2011 and create a dataset.


```python
df2=df[df.InvoiceDate>pd.to_datetime('2011-04-01 00:00:00')].copy()
df2=df2[df2.InvoiceDate<pd.to_datetime('2011-12-10 00:00:00')]
df2
```

## 1. c. Calculate the RFM values for each customer (by customer id). RFM represents:
###   R (Recency) − Recency should be calculated as the number of months before he or  she has made a purchase from the online store. If he/she made a purchase in the month of December 2011, then the Recency should be 0. If purchase is made in November 2011 then Recency should be 1 and so on and so forth.
###   F (Frequency) − Number of invoices by the customer from 01/04/ 2011 and 09/12/2011.
###   M (Monetary Value) − Total spend by the customer from 01/04/ 2011 and 9/12/2011.

```python

date = pd.to_datetime('2011-12-10 00:00:00')
recency_df = df2.groupby('CustomerID').agg({'InvoiceDate': lambda x: (date - x.max()).days // 30}).rename(columns={'InvoiceDate': 'Recency'})
frequency_df = df2.groupby('CustomerID').agg({'InvoiceNo': 'nunique'}).rename(columns={'InvoiceNo': 'Frequency'})
monetary_df = df2.groupby('CustomerID').agg({'Total': 'sum'}).rename(columns={'Total': 'Monetary'})
rfm_df = recency_df.merge(frequency_df, on='CustomerID').merge(monetary_df, on='CustomerID')
rfm_df
```

## 5. values for each customer.
## a. Use the elbow method to identify how many customer segments exist, using the RFM

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = rfm_df.loc[:,['Recency','Frequency','Monetary']]

sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data)
    data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python

kmeans = KMeans(n_clusters=3, random_state=42)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_df[['Recency','Frequency','Monetary']])

print(df2)
```

```python
cluster_summary = rfm_df.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'Cluster': 'count'
}).rename(columns={'Cluster': 'Num_Customers'})
cluster_summary
```

## 6. from sklearn.cluster import KMeans
### a. Plot the clusters in a scatter plot and mark each segment differently using lmplot.
### b. Print the cluster centers of each customer segment and explain them intuitively.
### c. Create the customer segments with Agglomerative algorithm by using number of clusters is suggested by elbow method.

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.lmplot(data=rfm_df.reset_index(), x='Recency', y='Frequency', hue='Cluster', palette='Set1', markers=["o", "s", "D"], height=15)
plt.show()
```

```python

cluster_centers = kmeans.cluster_centers_


centers_df = pd.DataFrame(cluster_centers, columns=['Recency', 'Frequency', 'Monetary'])
centers_df.index.name = 'Cluster'


print(centers_df)
```

```python
from sklearn.cluster import AgglomerativeClustering

agglo = AgglomerativeClustering(n_clusters=3)
rfm_df['Agglo_Cluster'] = agglo.fit_predict(rfm_df[['Recency', 'Frequency', 'Monetary']])
rfm_df
```

```python
cluster_summary2 = rfm_df.groupby('Agglo_Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'Agglo_Cluster': 'count'
}).rename(columns={'Agglo_Cluster': 'Num_Customers'})
cluster_summary2
```
